```python
import pandas as pd
import numpy as np
```

```python
employee = pd.read_csv('D:\\data science\\csv\\employee.csv')
```

```python
employee.head()
```

```python
employee.describe(include='all')
```

Few of the columns contain all similar values

```python
print(employee.EmployeeCount.unique())
print(employee.EmployeeNumber.unique())
print(employee.Over18.unique())
print(employee.StandardHours.unique())
```

Due to single value for some columns and id variable we drop them

```python
employee = employee.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'], axis=1)
```

```python
employee.columns
```

Do we have NAs in the data?

```python
employee.isna().sum()
```

```python
#replacing our Attrition output by integer constants
employee.loc[employee['Attrition']=='No','Attrition'] = 0
employee.loc[employee['Attrition']=='Yes','Attrition'] = 1
```

```python
#employee numeric columns
employee_num = employee[employee.select_dtypes(include=np.number).columns.tolist()]
```

```python
employee_num.info()
```

```python
employee_num.head()
```

```python
employee_dummies = pd.get_dummies(employee[employee.select_dtypes(include=['object']).columns.tolist()])
```

```python
employee_dummies.head()
```

```python
employee_combined = pd.concat([employee_num,employee_dummies], axis=1)
```

```python
employee_combined.head()
```

```python
#Dividing the data into train and test dataset
from sklearn.model_selection import train_test_split

train_x = employee_combined.drop(['Attrition'], axis=1)

train_y = employee_combined['Attrition']

#train test split
X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.3, random_state=20)
```

```python
#Import the classifier model
from sklearn import tree

dt = tree.DecisionTreeClassifier()

#train the model using training data
dt.fit(X_train, y_train)

#predict the response for test dataset
y_pred = dt.predict(X_test)
```

```python
#import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
print('Accuracy:', metrics.accuracy_score(y_test,y_pred))
print('Precision:', metrics.precision_score(y_test,y_pred))
print('Recall:', metrics.recall_score(y_test,y_pred))
```

Random Search

```python
from time import time
from scipy.stats import randint as sp_randint
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
#Utility function to report the best scores
def report(results, n_top=3):
    for i in range(1, n_top+1):
        candidates = np.flatnonzero(results['rank_test_score']== i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(results['mean_test_score'][candidate],results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
            
```

```python
list(range(1, 2))
```

```python
#specify parameters and distributions to sample from
param_dist = {'max_depth': [3,None],
              'max_features' : sp_randint(1,11),
              'min_samples_split' : sp_randint(2,11),
              'bootstrap' : [True,False],
              'criterion' : ['gini','entropy']}
```

```python
#run randomzied search
n_iter_search = 20
random_search = RandomizedSearchCV(rf, param_distributions=param_dist, n_iter=n_iter_search, cv=5)
start = time()
random_search.fit(X_train, y_train)
print("Randomized Search took %.2f seconds for %d candidates" " parameter settings." % ((time()-start), n_iter_search))
report(random_search.cv_results_)
```

```python
random_search.cv_results_
```

```python
len(random_search.cv_results_['rank_test_score'])
```

Grid Search

```python
#now use a full grid over all parameters
param_grid = {"max_depth": [3,None],
             "max_features": [1,3,10],
             "min_samples_split": [2,3,10],
             "bootstrap": [True,False],
             "criterion": ["gini","entropy"]}
```

```python
#run grid search
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5)
start=time()
grid_search.fit(X_train, y_train)

print("Grid search took %.2f seconds for %d candidate parameter settings" %(time()-start, len(grid_search.cv_results_['params'])))
report(grid_search.cv_results_)
```

```python
#now run the random forest with best parameters from the seracj

#build the classifier with the best of random search
rf_random_search = RandomForestClassifier(n_estimators=100, bootstrap=True, criterion="gini",
                                         max_features=10, min_samples_split=10)

#train the model using training sets with the best of Random Search
rf_random_search.fit(X_train, y_train)

#predict the response for test dataset
y_pred = rf_random_search.predict(X_test)
```

```python
#Print model performance metrics
print("Accuracy using Random Search:", metrics.accuracy_score(y_test,y_pred))
print("Precision using Random Search:", metrics.precision_score(y_test,y_pred))
print("Recall using Random Search:", metrics.recall_score(y_test,y_pred))
```

```python
#Build a classifier with the best of grid search
rf_grid_search = RandomForestClassifier(n_estimators=100, bootstrap=False, criterion="gini",
                                       max_features=10, min_samples_split=2)

#train the model
rf_grid_search.fit(X_train,y_train)

#predict
y_pred = rf_grid_search.predict(X_test)
```

```python
#print model performance metrics
print("Accuracy using Grid Search:", metrics.accuracy_score(y_test,y_pred))
print("Precision using Grid Search:", metrics.precision_score(y_test,y_pred))
print("Recall using Grid Search:", metrics.recall_score(y_test,y_pred))
```
