#### 5. Create a function titanic_barchart() which takes in categorical features as argument and returns a bar plot with distribution of categories for people who survived vs people who died.

#### 7. Extract the title of each passanger name using str.extract() and create a new feature called Title that contains the initials of each name such as Mr., Miss., Lady and Master.
###### Hint: Use Regular expression ([A-Za-z]+)\. 

```python
###Paste your solution here###
dataframe=[df_train,df_test]
def Title_Field_imputation(df_name):
    for k in df_name:
        k['Title']=k.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
        k['Title'] = k['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
        k['Title'] = k['Title'].replace('Mlle', 'Miss')
        k['Title'] = k['Title'].replace('Ms', 'Miss')
        k['Title'] = k['Title'].replace('Mme', 'Mrs')
        print(k.head())

Title_Field_imputation(dataframe)
```

#### 10. Build multiple classifiers such as logistic, decision trees, random forest and svm and predict using the model with best performance on the test data, report key metrics such as precison, recall, accuracy and confusion matrix to support your decision

```python
###Paste your solution here###
Numerical_data_cols = []
Categorical_data_cols = [] 
dummy_cat = []
num_df = []
cnt=0
#Dummies
for i in dataframe:
    Numerical_data_cols.append(i.select_dtypes(include=[np.number]).columns.tolist())
    Categorical_data_cols.append(i.select_dtypes(include=['object']).columns.tolist())
    Categorical_data_cols[cnt].remove('Ticket')
    Categorical_data_cols[cnt].remove('Name')
    dummy_cat.append(pd.get_dummies(i[Categorical_data_cols[cnt]]))
    num_df.append(i[Numerical_data_cols[cnt]])
    cnt+=1
df_train_combine = pd.concat([dummy_cat[0],num_df[0]],axis=1)

df_test_combine = pd.concat([dummy_cat[1],num_df[1]],axis=1)
X_train = df_train_combine.drop(['Survived'], axis=1)
y_train = df_train_combine['Survived']


y_test_df = pd.read_csv("D:\\data science\\csv\\titanic\\gender_submission.csv")
y_test = y_test_df.Survived
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

Lmre = LogisticRegression()
Lmre.fit(X_train,y_train)

prediction = Lmre.predict(df_test_combine)

from sklearn import metrics
print('Accuracy:', metrics.accuracy_score(y_test,prediction))
print('Precision:', metrics.precision_score(y_test,prediction))
print('Recall:', metrics.recall_score(y_test,prediction))
print('Confusion Matrix :\n', metrics.confusion_matrix(y_test,prediction))
```

### Decision Tree

```python
from sklearn import tree

dt = tree.DecisionTreeClassifier()

#train the model using training data
dt.fit(X_train, y_train)

#predict the response for test dataset
prediction = dt.predict(df_test_combine)

from sklearn import metrics
print('Accuracy:', metrics.accuracy_score(y_test,prediction))
print('Precision:', metrics.precision_score(y_test,prediction))
print('Recall:', metrics.recall_score(y_test,prediction))
print('Confusion Matrix :\n', metrics.confusion_matrix(y_test,prediction))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
#build a classifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train,y_train)
prediction = rf.predict(df_test_combine)
from sklearn import metrics
print('Accuracy:', metrics.accuracy_score(y_test,prediction))
print('Precision:', metrics.precision_score(y_test,prediction))
print('Recall:', metrics.recall_score(y_test,prediction))
print('Confusion Matrix :\n', metrics.confusion_matrix(y_test,prediction))
```
